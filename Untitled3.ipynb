```python
train_path = r'C:\Users\user\Downloads\Pikky_Assignment\Rice_and_Curry\train'
valid_path = r'C:\Users\user\Downloads\Pikky_Assignment\Rice_and_Curry\valid'
test_path = r'C:\Users\user\Downloads\Pikky_Assignment\Rice_and_Curry\test'
```

> import required libraries

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten, BatchNormalization, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline
```

>Now, we use Keras' ImageDataGenerator class to create batches of data from the train, valid, and test directories.

```python
train_batches = ImageDataGenerator(preprocessing_function = tf.keras.applications.vgg16.preprocess_input)\
.flow_from_directory(directory = train_path,target_size =(224,224), classes = ['curry','rice'], batch_size = 10)

valid_batches = ImageDataGenerator(preprocessing_function = tf.keras.applications.vgg16.preprocess_input) \
.flow_from_directory(directory= valid_path, target_size=(224,224), classes =['curry','rice'], batch_size= 10)

test_batches = ImageDataGenerator(preprocessing_function = tf.keras.applications.vgg16.preprocess_input)\
.flow_from_directory(directory = test_path, target_size = (224,224), classes =['curry', 'rice'], batch_size = 10, shuffle = False)
```

```python
imgs, labels = next(train_batches)
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20)) #10 images will be displayed
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(imgs)
print(labels)
```

**Note: [1,0] is used for Dosa i.e curry in our case and [0,1] is used for Biryani "Rice".**

```python
model = Sequential([
    Conv2D(32, (3,3), activation = 'relu', padding ='same', input_shape = (224,224,3)), #padding= 'same' gives zero-padding
    MaxPool2D((2,2), strides = 2), #strides =2, mean cutting the images into half
    Conv2D(64, (3,3), activation = 'relu', padding = 'same'),
    MaxPool2D((2,2), strides = 2),
    Flatten(),
    Dense(2, activation = 'softmax') #output
])
```

```python
model.summary()
```

```python
model.compile(optimizer ='Adam', loss ='categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x=train_batches,
    steps_per_epoch=len(train_batches),
    validation_data=valid_batches,
    validation_steps=len(valid_batches),
    epochs=10,
    verbose=2
)
```

```python
test_imgs, test_labels = next(test_batches)
```

```python
plotImages(test_imgs)
print(test_labels)
```

```python
predictions = model.predict(x=test_batches, steps=len(test_batches), verbose=0)
```

```python
import numpy as np
np.round(predictions)
```

>Confusion matrix to demonstrate results

```python
cm = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
```
